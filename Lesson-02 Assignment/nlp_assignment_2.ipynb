#### 1.1 Random Choose Method to get optimal k and b

```python
import random
```

```python
from sklearn.datasets import load_boston
```

```python
data = load_boston()
```

```python
x,y = data["data"],data["target"]
```

```python
x_rm = x[:,5]
```

```python
list(x_rm)
```

```python
k = random.random()
```

```python
b = random.random()
```

```python
k,b
```

```python
def price(r,k,b):
    return k * r + b 
```

```python
def loss(y,y_hat):
    return sum((y_i - y_hat_i) ** 2 for y_i,y_hat_i in zip(list(y),list(y_hat))) / len(list(y))
```

```python
trying_times = 2000
```

```python
min_loss = float("inf")
```

```python
for i in range(trying_times):
    k = random.random() * 200 - 100
    b = random.random() * 200 - 100
    y_hat = [price(r,k,b) for r in list(x_rm)]
    current_loss = loss(y,y_hat)
    if current_loss < min_loss:
        min_loss = current_loss
        best_k,best_b = k,b
        print('When time is : {}, get best_k: {} best_b: {}, and the loss is: {}'.format(i, best_k, best_b, min_loss))
```

#### 1.2 Supervised Direction to get optimal k and b

```python
direction = [(1,1),(-1,-1),(1,-1),(-1,1)]
```

```python
scalar = 0.1
```

```python
min_loss = float("inf")
```

```python
best_k = random.random() * 200 - 100
```

```python
best_b = random.random() * 200 - 100
```

```python
for i in range(trying_times):
    k_direction,b_direction = random.choice(direction)
    current_k ,current_b = best_k + k_direction * scalar,best_b + b_direction * scalar
    y_hat = [price(r,current_k,current_b) for r in x_rm]
    current_loss = loss(y,y_hat)
    if current_loss < min_loss:
        min_loss = current_loss
        best_k,best_b = current_k,current_b
        k_direction,b_direction = k_direction,b_direction
        print('When time is : {}, get best_k: {} best_b: {}, and the loss is: {}'.format(i, best_k, best_b, min_loss))
    else:
        k_direction,b_direction = random.choice(direction)
```

#### 1.3 Gradient Descent to get optimal k and b

```python
def partial_k(x,y,y_hat):
    gradient = 0
    n = len(y)
    for x_i,y_i,y_hat_i in zip(list(x),list(y),list(y_hat)):
        gradient += (y_i - y_hat_i) * x_i
    return -2 / n * gradient
```

```python
def partial_b(y,y_hat):
    gradient = 0
    n = len(y)
    for y_i,y_hat_i in zip(list(y),list(y_hat)):
        gradient += (y_i - y_hat_i) 
    return -2 / n * gradient
```

```python
current_k = random.random() * 200 -100
```

```python
current_k
```

```python
current_b = random.random() * 200 -100
```

```python
learning_rate = 0.01
```

```python
min_loss = float("inf")
```

```python
for i in range(trying_times):
    price_by_k_and_b = [price(r,current_k,current_b) for r in x_rm]
    current_loss = loss(y,price_by_k_and_b)
    if current_loss < min_loss :
        min_loss = current_loss
        if i % 50 == 0:
            print(
                'When time is : {}, get best_k: {} best_b: {}, and the loss is: {}'.format(i, best_k, best_b, min_loss))
    gradient_k = partial_k(x_rm,y,price_by_k_and_b)
    gradient_b = partial_b(y,price_by_k_and_b)
    current_k = current_k - gradient_k * learning_rate
    current_b = current_b - gradient_b * learning_rate
    
```
